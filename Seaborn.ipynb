**Boxplots**

A boxplot is a standardized way of displaying the distribution of data based on a five number("minimum",first quartile (Q1), median, third quartile (Q3), and "maximum"). It can tell about your outliers and what their values are. it can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

# univariate

```python
sns.boxplot(x=tips['tip'])
# outliers for miimun value  and outliers maximun malue and outliers and also a intercortail range
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='day', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=tips)
```

**Violin Plots**

violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot.

The unquestionate advantage of the violin plot cover the box plot is that aside from showing the abovementioned statistics It also shows the entire distribution of the data

Viloin plots are often used to compare the distribution of a given variable across some categories

```python
sns.violinplot(x=tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='day', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split='True', data=tips)
```

**Heatmaps**

A heat map is a graphical representation of data where the individual values contained in a matrix are represented are colos.
It is between the two categorical data, but sometime it is between dicrete numerical data

```python
flights = sns.load_dataset('flights')
# The dataset is basically related to us airline
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
#This is known as color bar, which shows tje color scalling
```

```python
sns.heatmap(x, cbar=False, linewidth=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidth=0.5, annot=True, fmt='d', cmap='winter')
```

**Clustermap**

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score=0, annot=True, row_cluster=False, metric='correlation' )
```

```python
sns.clustermap(x, z_score=0, annot=True, col_cluster=False, metric='correlation' )
```

**Joinplot**

Jointplot creates a multi-panel figure that shows the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

we can use jointplots in the 4 manners

1. ScatterPlots
2. Hexbin plots
3. kidePlots
4. Reglots

```python
tips=sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips)
# hex basically 2d histogram, and the higher has darker shade
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
# univariate and kde is 2d plot
# This is known as pounto, it is basically 2d pdf
# The Darker shade higher point on the peak and light shade lower point on the peak
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

**Pairplots**

Pairplots are used to plot multiple pairwise bivariate distribution in a dataset. This creates a matrix of axes and shows the relationship for each other pair of columns in a DataFrame. By default, It also draws the univariate distribution of each variable on the diagonal Axes.

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
# Histogram, Its a relationship of varaible with it self, and other plot are scatter plot because they are bivariate relationship between two different variables
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.pairplot(iris, hue='species', palette='rainbow')
```

**Displots**

Displots are used to do univariate analysis on a contineous data.

Displot combines 3 things in one graph.

1. Histogram
2. Kernel Density Enstimation (PDF)
3. Rugplot

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.distplot(titanic['age'], rug=True)
```

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'], hist=False, rug=True)
```

```python
sns.distplot(titanic['age'], bins=30)
```
